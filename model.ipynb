```python
from roboflow import Roboflow
import supervision as sv
import cv2

# Initialize Roboflow
rf = Roboflow(api_key="XBkxfvTxWHoyX6BYi1t0")
project = rf.workspace().project("tennis-court-keypoints")
model = project.version(1).model

# Run inference on your image
result = model.predict("your_image.jpg", confidence=40, overlap=30).json()

# Extract prediction labels
labels = [item["class"] for item in result["predictions"]]

# Convert to supervision Detections object
detections = sv.Detections.from_roboflow(result)

# Create annotators
label_annotator = sv.LabelAnnotator()
box_annotator = sv.BoxAnnotator()  # ✅ This was missing

# Load image with OpenCV
image = cv2.imread("input_videos/image.png")

# Annotate image with bounding boxes and labels
annotated_image = box_annotator.annotate(scene=image, detections=detections)
annotated_image = label_annotator.annotate(scene=annotated_image, detections=detections, labels=labels)

# Display the annotated image
sv.plot_image(image=annotated_image, size=(16, 16))
```
